Q1. Explain GET and POST methods.

Ans: The GET and POST methods are two HTTP (Hypertext Transfer Protocol) request methods used for different purposes in web development.
GET Method:
Purpose: The GET method is used to request data from a specified resource.
Data Transmission: Data is appended to the URL in the form of parameters. This means that the data is visible in the URL, which makes it less secure for transmitting sensitive information.
POST Method:
Purpose: The POST method is used to submit data to be processed to a specified resource. It's commonly used when dealing with forms and sensitive data.
Data Transmission: Data is sent in the body of the HTTP request, rather than appended to the URL. This makes it more secure for transmitting sensitive information.

Q2. Why is request used in Flask?

Ans: In Flask, the request object is used to access incoming request data, whether it's data submitted through a form, query parameters in a URL, or any other type of data sent by a client to the server. The request object provides a convenient way to interact with this incoming data within your Flask application.

Q3. Why is redirect() used in Flask?

Ans: In Flask, the redirect() function is used to perform an HTTP redirect. It's a convenient way to redirect the user's browser to a different endpoint (URL) within your Flask application or to an external URL.

Q4. What are templates in Flask? Why is the render_template() function used?

Ans: In Flask, templates are used to separate the structure and layout of the HTML content from the Python code that generates it. Templates allow you to build dynamic web pages by embedding variables, control structures, and other logic into HTML, making it easier to manage and maintain your code.
The render_template() function in Flask is used to render templates and generate HTML content dynamically. It takes the name of the template file as an argument and, optionally, additional variables that should be passed to the template.
