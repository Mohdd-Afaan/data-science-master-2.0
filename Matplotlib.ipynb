Q2: What is a scatter plot?


Ans: A scatter plot is a type of data visualization that displays individual data points on a two-dimensional graph. Each point on the graph represents the values of two different variables. The x-axis typically represents one variable, and the y-axis represents the other variabl

Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.xlabel("This is my x-label")
plt.ylabel("This is my y-lablel")
plt.scatter(x,y)
```

Q3: Why is the subplot() function used?

Ans:
The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. This function is particularly useful when you want to display multiple plots or visualizations in a single figure, arranging them in rows and columns.

Draw four line plots using the subplot() function.
Use the following data:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])


plt.subplot(2,2,1)
plt.plot(x, np.array([0, 100, 200, 300, 400, 500]))
plt.title("Line 1")

plt.subplot(2,2,2)
plt.plot(x, np.array([50, 20, 40, 20, 60, 70]))
plt.title("Line 2")

plt.subplot(2,2,3)
plt.plot(x, np.array([10, 20, 30, 40, 50, 60]))
plt.title("Line 3")

plt.subplot(2,2,4)
plt.plot (x, np.array([200, 350, 250, 550, 450, 150]))
plt.title("Line 4")

plt.tight_layout()

plt.show()
```

Q4: What is a bar plot? Why is it used?

Ans: A bar plot (or bar chart) is a data visualization that uses rectangular bars to represent the values of different categories or groups. The length of each bar is proportional to the value it represents. Bar plots are commonly used for visualizing and comparing the quantities of different categories or groups.

Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Profit of Companies')

plt.barh(company,profit)
```

Q5: What is a box plot? Why is it used?

Ans: A box plot is a statistical data visualization that provides a graphical summary of the distribution of a dataset. It displays key statistics, such as the median, quartiles, and potential outliers, in a compact and easily interpretable way. Box plots are commonly used for the following purposes:
1) Box plots provide a concise summary of the distribution of a dataset, including the central tendency (median), spread (interquartile range), and potential outliers.
2) Outliers, or data points that fall significantly outside the overall pattern of the data, can be easily identified in a box plot. Outliers are typically represented as individual points beyond the "whiskers" of the plot.
4) Box plots are useful for comparing the distributions of multiple datasets or groups. Multiple box plots can be drawn side by side for visual comparison

Using the following data plot a box plot.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])

plt.xlabel('Boxes')
plt.ylabel('Values')
plt.title('Box Plot Example')
```
