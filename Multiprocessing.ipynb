Q2. What are the differences between multiprocessing and multithreading?

Ans: Multiprocessing and multithreading are both techniques used to achive concurrency in a program, allowing multiple tasks to be executed simultaneously. However, they differ in their approach and usage:
1)Process vs Thread:
  Multiprocessing: In multiprocessing, the program is divided into multiple processes, each with its own space. These run independently and can be executed on different cpu cores.
  Multithreading: In multithreading, it involves creating multiple threads within a single processor. Threads within a process shares memory, resources which make them lightweight in terms of resource consumption
2)Parallelism vs Concurrency:
  Multiprocessing: It is well suited for achieving true parallelism on multi-core cpu. Each process runs independently, taking full advantage of available CPU cores. This is particularly used for CPu bound tasks that can be executed in parallel.  
  Multithreading: It is often used to achive concurrency, where multiple threads execute tasks concurrently within a single processor

Q3. Write a python code to create a process using the multiprocessing module.

```python
import multiprocessing
from multiprocessing import process
import time
def mpf():
    print("hi")
    time.sleep(1)
    print("after 1 second")
if __name__ == "__main__":
    processes = []
    for _ in range(10):
        processes.append(multiprocessing.Process(target=mpf))
    for i in processes:
        i.start()
        i.join()
```

Q4. What is a multiprocessing pool in python? Why is it used?

Ans: A multiprocessing pool is provided by multiprocessing module for creating  
     and managing a group of worker processes. It is used to distribute and parallelize tasks across multiple processes efficiently. It is used to distribute and parallelize tasks across multiple processes efficiently.
     It is used for a variety of reasons:
     1) Parallel execution: It enables you to execute multiple tasks concurrently taking advantage of multiple cpu cores and improving the performance of cpu bound tasks.
     2) Simplified parallelism: It abstract away much of the complexity of managing individual processes and communication between them. This simplifies the development of other programs.
     3) Load balancing: A distribute task among available worker processes, Ensuring that the work is evenly distributed and making efficient use of available resources.

Q6. Write a python program to create 4 processes, each process should print a different number using the multiprocessing module in python.
